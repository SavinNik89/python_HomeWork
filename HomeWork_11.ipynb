```python
from sympy import Symbol, diff, oo, Interval
from sympy import is_increasing, is_decreasing
from sympy import plot, solve
x = Symbol('x', real=True)
f = x**2 - 6*abs(x) + 8
f
```

```python
ans = solve(f)
ans
```

```python
plot(f)
```

```python
derivate = diff(f, x)
derivate
```

```python
plot(derivate)
```

```python
der_ans = solve(derivate)
der_ans
# почему не выводит отрицательный корень не понятно, видимо проблема в том, что функция содержит модуль?
# можете объяснить в обратной связи?
```

```python
if is_increasing (f, Interval(-oo, -3)):
    print(f'Функция возрастает в {Interval(-oo, -3)}')
else:
    print(f'Функция убывает в {Interval(-oo, -3)}')

if is_increasing (f, Interval(-3, 0)):
    print(f'Функция возрастает в {Interval(-3, 0)}')
else:
    print(f'Функция убывает в {Interval(-3, 0)}')
if is_increasing (f, Interval(0, 3)):
    print(f'Функция возрастает в {Interval(0, 3)}')
else:
    print(f'Функция убывает в {Interval(0, 3)}')
if is_increasing (f, Interval(3, +oo)):
    print(f'Функция возрастает в {Interval(3, +oo)}')
else:
    print(f'Функция убывает в {Interval(3, +oo)}')
```

```python
extremum = f.subs (x, der_ans[0])
extremum
```
