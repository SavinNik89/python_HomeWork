```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cwd = os.getcwd()
pd_df = pd.read_csv(os.path.join(cwd, "company_sales_data.csv"))
pd_df.head()
```

```python
month_list = pd_df['month_number']
facecream_list = pd_df['facecream']
facewash_list = pd_df['facewash']
toothpaste_list = pd_df['toothpaste']
bathingsoap_list = pd_df['bathingsoap']
shampoo_list = pd_df['shampoo']
moisturizer_list = pd_df['moisturizer']
total_units_list = pd_df['total_units']
total_profit_list = pd_df['total_profit']
```

```python
fig = plt.figure(figsize=(15, 15)) 
ax = fig.subplots()

ax.plot(month_list, facecream_list, 'b', label = 'facecream')
ax.plot(month_list, facewash_list, 'r', label = 'facewash')
ax.plot(month_list, toothpaste_list, 'y', label = 'toothpaste')
ax.plot(month_list, bathingsoap_list, 'g', label = 'bathingsoap')
ax.plot(month_list, shampoo_list, 'k', label = 'shampoo')
ax.plot(month_list, moisturizer_list, 'm', label = 'moisturizer')

plt.xticks(month_list)
plt.yticks(range(0, 15000, 1000))
plt.title('Количество проданных товаров (шт):')
plt.ylabel('Продано (шт)')
plt.xlabel('Месяц')
plt.legend(loc = 'upper left')
```

```python
fig2 = plt.figure(figsize=(8, 8)) 
ax2 = fig2.subplots()
ax2.bar(month_list, bathingsoap_list, color = 'lime')
ax2.grid(axis = 'y')

plt.xticks(month_list)
plt.yticks(range(0, 15000, 1000))
plt.title('Объем продаж bathingsoap:')
plt.ylabel('Продано (шт)')
plt.xlabel('Месяц')
```

```python
all_sale = [sum(facewash_list), sum(facecream_list), sum(toothpaste_list), sum(bathingsoap_list), sum(shampoo_list), sum(moisturizer_list)]
label_all_sale = ['facewash', 'facecream', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

fig3 = plt.figure(figsize=(8, 8)) 
ax3 = fig3.subplots()
                                                                               
ax3.pie(all_sale, labels = label_all_sale, autopct='%.2f%%')


```
